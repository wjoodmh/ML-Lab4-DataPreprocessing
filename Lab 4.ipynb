```python
# ================================
# Step 1: Import Libraries
# ================================
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA

# ================================
# Step 2: Load Dataset
# ================================
df = pd.read_csv('custom_large_dataset.csv')

print("First 5 rows:")
print(df.head())


# ================================
# Task 1: Data Quality Issues
# ================================
print("\nMissing Values:")
print(df.isnull().sum())

print("\nDuplicate Rows:")
print(df.duplicated().sum())


# ================================
# Task 2: Handle Missing Values
# ================================
# استخدام mean لأن البيانات رقمية

df.fillna(df.mean(), inplace=True)

print("\nAfter Handling Missing Values:")
print(df.isnull().sum())


# ================================
# Task 3: Detect & Handle Outliers (IQR)
# ================================
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

df_clean = df[~((df < lower) | (df > upper)).any(axis=1)]

print("\nShape after removing outliers:", df_clean.shape)


# ================================
# Task 4: Normalization
# ================================

# Min-Max
df_minmax = (df_clean - df_clean.min()) / (df_clean.max() - df_clean.min())

print("\nMin-Max Normalization:")
print(df_minmax.head())

# Z-score
df_zscore = (df_clean - df_clean.mean()) / df_clean.std()

print("\nZ-score Normalization:")
print(df_zscore.head())


# ================================
# Task 5: PCA (only if correlated)
# ================================

# حساب ال correlation
corr = df_clean.corr()

print("\nCorrelation Matrix:")
print(corr)

# إذا فيه ارتباط عالي نطبق PCA
if (abs(corr) > 0.5).any().any():
    print("\nApplying PCA...")

    pca = PCA(n_components=2)
    pca_result = pca.fit_transform(df_clean)

    print("\nPCA Result (first rows):")
    print(pca_result[:5])
else:
    print("\nNo strong correlation, PCA not applied.")
```
